```python
# Load project data from CSV
project = pd.read_csv("project.csv")

project
```

```python
# Convert cost column into list
cost_list = project["Cost"].tolist()

# Loop to replace missing values
for i in range(len(cost_list)):
    if pd.isna(cost_list[i]):  # check missing value
        avg = np.nanmean(cost_list[:i])  # running average
        cost_list[i] = avg  # replace with avg

# Update dataframe
project["Cost"] = cost_list

# Show updated dataframe
project
```

```python
# Split Name into First Name and Last Name
employee[["First Name", "Last Name"]] = employee["Name"].str.split(" ", expand=True)

# Drop original Name column
employee.drop("Name", axis=1, inplace=True)

# Show updated dataframe
employee
```

```python
# Load all datasets
employee = pd.read_csv("employee.csv")
seniority = pd.read_csv("seniority.csv")
project = pd.read_csv("project.csv")
```

```python
# Merge employee and seniority
final = pd.merge(employee, seniority, on="ID")

# Merge with project
final = pd.merge(final, project, on="ID")

# Display final dataframe
final
```

```python
# Add bonus column (5% if project is Finished)
final["Bonus"] = np.where(final["Status"] == "Finished",
                          final["Cost"] * 0.05,
                          0)

# Show updated dataframe
final
```

```python
# Demote designation level by 1 where project failed
final.loc[final["Status"] == "Failed", "Designation Level"] += 1

# Remove employees with designation level > 4
final = final[final["Designation Level"] <= 4]

# Show updated dataframe
final
```

```python
# Add Mr. and Mrs. to Name
final["Name"] = np.where(final["Gender"] == "M",
                        "Mr. " + final["Name"],
                        "Mrs. " + final["Name"])

# Drop Gender column
final.drop("Gender", axis=1, inplace=True)

# Show updated dataframe
final
```

```python
# Promote designation level for age > 29
final.loc[final["Age"] > 29, "Designation Level"] -= 1

# Show updated dataframe
final
```

```python
# Calculate total project cost per employee
total_proj_cost = final.groupby(["ID", "Name"])["Cost"].sum().reset_index()

# Rename column
total_proj_cost.rename(columns={"Cost": "Total Cost"}, inplace=True)

# Show result
total_proj_cost
```
